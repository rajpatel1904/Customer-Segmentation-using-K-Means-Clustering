```python
import numpy as np
import pandas as pd
import seaborn as sea
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

Data Collection

```python
customers=pd.read_csv('Mall_Customers.csv')
```

working with Annual Income and Spending Score Column

```python
S=customers.iloc[:,[3,4]].values
```

```python
print(S)
```

choosing the number of clusters centers


```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(S)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```


Training the K-Means Clustering model

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=42)
Output=kmeans.fit_predict(S)
print(Output)
```

```python


import matplotlib.pyplot as plt
# Visualizing all the clusters
plt.figure(figsize=(8,8))
plt.scatter(S[Output==0,0],S[Output==0,1],s=50,c='green',label='Cluster 1')
plt.scatter(S[Output==1,0],S[Output==1,1],s=50,c='red',label='Cluster 2')
plt.scatter(S[Output==2,0],S[Output==2,1],s=50,c='yellow',label='Cluster 3')
plt.scatter(S[Output==3,0],S[Output==3,1],s=50,c='black',label='Cluster 4')
plt.scatter(S[Output==4,0],S[Output==4,1],s=50,c='blue',label='Cluster 5')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='cyan',label='Centroids')

plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
